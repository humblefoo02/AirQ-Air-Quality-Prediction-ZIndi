```python
import pandas as pd
```

```python
df = pd.read_csv('Train.csv')
df.head()
```

```python
X = df.drop('pm2_5', axis=1)
y = df['pm2_5']
```

```python
from lightgbm import LGBMRegressor
model = LGBMRegressor()
```

```python
X = X.drop('id', axis=1)
X = X.drop('site_id', axis=1)
X = X.drop('city', axis=1)
X = X.drop('country', axis=1)
X = X.drop('date', axis=1)
```

```python
model.fit(X, y)
```

```python
X_test = pd.read_csv('Test.csv')
X_test = X_test.drop('id', axis=1)
X_test = X_test.drop('site_id', axis=1)
X_test = X_test.drop('city', axis=1)
X_test = X_test.drop('country', axis=1)
X_test = X_test.drop('date', axis=1)
```

```python
pred = model.predict(X_test)
```

```python
X_test1 = pd.read_csv('Test.csv')
sub = pd.DataFrame({'id': X_test1['id'], 'pm2_5': pred})
```

```python
sub.to_csv('submission_lgbm_op1.csv', index = False)
```
